##Previsões de séries temporais com Arima

Arima é uma ténicas muito utilizadas para previsões de série temporais. Ela requer que os dados estejam estecionário. Caso não estejam estacionários, a própria técnica faz a diferenciação e transforma os dados em estacionários.

Instalando e carregando a biblioteca

````python
!pip install pmdarima
from pmdarima.arima import auto_arima
````

Procurando o melhor modelo

````python
stepwise_model = auto_arima(ts_meses, start_p=1,start_q=1,start_d= 0, start_P=0, max_p=6, max_q=6, m=12, seasonal=True, trace=True, stepwise=False)
````

````python
print(stepwise_model.aic())
````

````python
train = ts_meses.loc['2014-01-01':'2017-12-31']
test = ts_meses.loc['2018-01-01':]
````

Treinando o modelo

````python
stepwise_model.fit(train)
````

Previsão de 12 meses

````python
future_forecast = stepwise_model.predict(n_periods=12)
````

````python
future_forecast
````

````python
future_forecast.sum()
````

````python
future_forecast = pd.DataFrame(future_forecast,index = test.index,columns=["DTOBITO"])
````

Comparando a previsão com o ocorrido!

````python
pd.concat([test,future_forecast],axis=1).plot(legend=False);
````

Inserindo a previsão no gráfico!

````python
plt.figure(figsize=(12,8))
plt.plot(train,label="Treino")
plt.plot(test,label="Teste")
plt.plot(future_forecast,label="Previsão")
plt.legend(loc = 'upper left')
plt.show()
````

##Métricas da previsão

````python
from sklearn.metrics import r2_score,mean_squared_error, mean_absolute_error, mean_absolute_percentage_error

r2 = r2_score(test,future_forecast)
MAE = mean_absolute_error( test, future_forecast )
MSE = mean_squared_error( test, future_forecast )
RMSE = mean_squared_error(test, future_forecast, squared=False )
MAPE = mean_absolute_percentage_error(test, future_forecast) * 100

r2
Dicionario = {
    'R²' : round( r2, 2 ),
    'MAE' : round(MAE, 2),
    'MSE' : round(MSE, 2),
    'RMSE' : round(RMSE, 2),
    'MAPE' : round(MAPE, 2)
}

pd.DataFrame( Dicionario, index=[0] ).transpose().rename( columns={0:'Valor da métrica'} )
````
